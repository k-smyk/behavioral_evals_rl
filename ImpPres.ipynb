```python
# installs and imports

!pip install transformers accelerate bitsandbytes sentencepiece datasets
from transformers import LlamaForCausalLM, LlamaTokenizer, AutoTokenizer
import torch
import datasets
from datasets import load_dataset
```

```python
# load the models

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-hf")
model = LlamaForCausalLM.from_pretrained("meta-llama/Llama-2-7b-hf",load_in_4bit=True,device_map="auto",bnb_4bit_use_double_quant=True,bnb_4bit_quant_type="nf4",bnb_4bit_compute_dtype=torch.float16)
tokenizer_chat = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf")
model_chat = LlamaForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",load_in_4bit=True,device_map="auto",bnb_4bit_use_double_quant=True,bnb_4bit_quant_type="nf4",bnb_4bit_compute_dtype=torch.float16)

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

model.eval()
model_chat.eval()
```

```python
# get presupposition datasets

pres_sets = ["cleft_existence", "both_presupposition", "change_of_state", "question_presupposition"]
pres_data = []
for name in pres_sets:
  ds = load_dataset("imppres",f"presupposition_{name}",split=name)
  ds = ds.train_test_split(test_size = 50, seed = 17)   # get random subset (size 50) of each set
  pres_data += [ds["test"]]

pres = datasets.concatenate_datasets(pres_data)
```

```python
# get implicature datasets

imp_sets = ["quantifiers","gradable_adjective","gradable_verb","numerals_2_3"]
imp_data = []
for name in imp_sets:
  ds = load_dataset("imppres",f"implicature_{name}",split=name)
  ds = ds.train_test_split(test_size = 50, seed = 17)
  imp_data += [ds["test"]]

imp = datasets.concatenate_datasets(imp_data)
```

```python
# log prob computation from HW2

def get_log_prob_of_completion(
        model,
        tokenizer,
        prompt,
        completion,
):
        """
        Convenience function for computing the log probability of a completion
        given a prompt. This is used to compute the log probability of the
        correct and incorrent labels given different trials for the different
        SuperGLUE tasks.
        """
        # tokenize the prompt and the completion separately
        # so as to access the logits for the completion only
        # when scoring the completion
        input_ids_completion = tokenizer(
                completion,
                return_tensors='pt'
        )['input_ids'].to(device)
        # truncate so as to fit into to maximal context window of gpt-2
        # which is 1024 tokens
        input_ids_prompt = tokenizer(
                prompt,
                return_tensors='pt',
                truncation=True,
                max_length=1024-input_ids_completion.shape[-1]
        )['input_ids'].to(device)
        # putting together the prompt and completion
        input_ids = torch.cat(
                (input_ids_prompt, input_ids_completion),
                dim=-1
        )
        # create attention mask and position ids
        attention_mask = (input_ids != tokenizer.eos_token_id).to(dtype=torch.int64)
        position_ids = attention_mask.cumsum(-1)-1
        # get the logits for the completion
        with torch.no_grad():
                out = model(
                        input_ids=input_ids,
                        attention_mask=attention_mask,
                        position_ids=position_ids
                )

        # get the logits of the completion
        # for that, make a tensor of the logits
        # for the completion only
        # in particular, we shift the indices by one to the left to access logits of the
        # actual sequence tokens
        logits_completion = out.logits[:, (input_ids_prompt.shape[-1]-1):-1]
        logits_completion = logits_completion.squeeze()
        # get the log probabilities for the completion
        log_probs = torch.nn.functional.log_softmax(
                logits_completion,
                dim=-1
        )
        # retrieve the logit corresponding to the actual completion tokens
        try:
                log_completion_tokens = log_probs.gather(
                        dim=-1,
                        index=input_ids_completion.squeeze().unsqueeze(-1)
                )
        except:
                log_completion_tokens = log_probs.gather(
                        dim=-1,
                        index=input_ids_completion.unsqueeze(-1)
                )
        completion_log_prob = torch.mean(
                log_completion_tokens
        ).cpu()

        return completion_log_prob
```

```python
import pandas as pd

# function for generating prompts and getting log probabilities

def evaluate(mod, tkn, ds):

  results = []

  for i in range(len(ds)):

    print(f"Sample {i}/{len(ds)}")

    sample = ds[i]

    prompt = f"{sample['premise']} {sample['hypothesis']}"    # prompt is a concatenation of the premise and hypothesis

    log_prob_entailment = get_log_prob_of_completion(
        model = mod,
        tokenizer = tkn,
        prompt = prompt,
        completion = "Entailment",
    )

    log_prob_neutral = get_log_prob_of_completion(
        model = mod,
        tokenizer = tkn,
        prompt = prompt,
        completion = "Neutral",
    )

    log_prob_contradiction = get_log_prob_of_completion(
        model = mod,
        tokenizer = tkn,
        prompt = prompt,
        completion = "Contradiction",
    )

    log_probs = [log_prob_entailment, log_prob_neutral, log_prob_contradiction]
    prediction = log_probs.index(max(log_probs))

    results += [[prompt, log_prob_entailment, log_prob_neutral, log_prob_contradiction, prediction]]

  df = pd.DataFrame(results)
  df.columns = ["prompt","entailment_prob","neutral_prob","contradiction_prob","prediction"]

  return df
```

```python
# run evaluation

pres_results = evaluate(model, tokenizer, pres)
pres_results_chat = evaluate(model_chat, tokenizer_chat, pres)
imp_results = evaluate(model, tokenizer, imp)
imp_results_chat = evaluate(model_chat, tokenizer_chat, imp)
```

```python
# save results and download csv files
from google.colab import files

pres_results_chat = pres_results_chat.drop(columns=["prompt"])
pres_results_chat.columns = ["entailment_prob_chat","neutral_prob_chat","contradiction_prob_chat","prediction_chat"]
imp_results_chat = imp_results_chat.drop(columns=["prompt"])
imp_results_chat.columns = ["entailment_prob_chat","neutral_prob_chat","contradiction_prob_chat","prediction_chat"]

# save datasets
pres_df = pd.DataFrame(pres)
pres_df = pd.concat([pres_df,pres_results,pres_results_chat], axis=1)
pres_df.to_csv("pres.csv")
files.download("pres.csv")

imp_df = pd.DataFrame(imp)
imp_df = pd.concat([imp_df,imp_results,imp_results_chat], axis=1)
imp_df.to_csv("imp.csv")
files.download("imp.csv")
```
