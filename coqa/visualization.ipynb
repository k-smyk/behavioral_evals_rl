```python
coqa_7b["Dataset"] = "coqa 7b"
coqa_7b_fs["Dataset"] = "coqa 7b few shot"
coqa_7b_chat["Dataset"] = "coqa 7b chat"
coqa_7b_chat_fs["Dataset"] = "coqa 7b chat few shot"

coqa_13b["Dataset"] = "coqa 13b"
coqa_13b_fs["Dataset"] = "coqa 13b few shot"
coqa_13b_chat["Dataset"] = "coqa 13b chat"
coqa_13b_chat_fs["Dataset"] = "coqa 13b chat few shot"
```

```python
coqa_7b["avg_log_prob"] = coqa_7b["log_prob"].mean()
coqa_7b_fs["avg_log_prob"] = coqa_7b_fs["log_prob"].mean()
coqa_7b_chat["avg_log_prob"] = coqa_7b_chat["log_prob"].mean()
coqa_7b_chat_fs["avg_log_prob"] = coqa_7b_chat_fs["log_prob"].mean()

coqa_13b["avg_log_prob"] = coqa_13b["log_prob"].mean()
coqa_13b_fs["avg_log_prob"] = coqa_13b_fs["log_prob"].mean()
coqa_13b_chat["avg_log_prob"] = coqa_13b_chat["log_prob"].mean()
coqa_13b_chat_fs["avg_log_prob"] = coqa_13b_chat_fs["log_prob"].mean()
```

```python
combined_df = pd.concat([df[['log_prob', 'Dataset']] for df in [coqa_7b,
                                                                coqa_7b_fs,
                                                                coqa_7b_chat,
                                                                coqa_7b_chat_fs,
                                                                coqa_13b,
                                                                coqa_13b_fs,
                                                                coqa_13b_chat,
                                                                coqa_13b_chat_fs]])

combined_df.reset_index(drop=True, inplace=True)

plt.figure(figsize=(10, 6))
sns.lineplot(data=combined_df, x=combined_df.index, y='log_prob', hue='Dataset')

plt.title('Log Probabilities Comparison')
plt.xlabel('Index')
plt.ylabel('Log Probability')
plt.show()
```

```python
combined_df = pd.concat([df[['avg_log_prob', 'Dataset']] for df in [coqa_7b,
                                                                coqa_7b_fs,
                                                                coqa_7b_chat,
                                                                coqa_7b_chat_fs,
                                                                coqa_13b,
                                                                coqa_13b_fs,
                                                                coqa_13b_chat,
                                                                coqa_13b_chat_fs]])
```

```python
combined_df.head()
```

```python
combined_df.drop_duplicates(ignore_index=True, inplace=True)
combined_df.head(10)
```

```python
max_avg_log_prob = combined_df['avg_log_prob'].max()
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(data=combined_df, x='Dataset', y='avg_log_prob', hue='Dataset')
plt.axhline(y=max_avg_log_prob, color='red', linestyle='--', label='Maximal Average Log Prob')
plt.title('Average Log Probabilities of Datasets')
plt.xlabel('Dataset')
plt.ylabel('Average Log Probability')
plt.legend()
plt.show()
```
